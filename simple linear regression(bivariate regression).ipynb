```python
#import all dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

# data preparation

```python
#import dataset
data=pd.read_csv("linear regression dataset.csv")
data.head()
```

```python
#check data type and get info about data 
data.info()
```

```python
#handling missing values
data=data.dropna()
```

```python
data.columns
```

```python
#extract dependent and independent variable
x=data.drop('y',axis=1)
y=data.drop('x',axis=1)
```

```python
#split data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# model fitting

```python
#fit model to splited data
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#predicted data
y_pred=model.predict(x_test)
```

```python
y_train_pred=model.predict(x_train)
```

# model evaluatiion

```python
#calculating error score r2 for test data
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('mean squared error:',mse)
print('r2 value:',r2)
```

```python
#calculating error score r2 for test data
mse_train=mean_squared_error(y_train,y_train_pred)
r2=r2_score(y_train,y_train_pred)
print('mean squared error:',mse)
print('r2 value:',r2)
```

```python
#plot data in graph
plt.figure(figsize=(15,8))
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,'r-')   #r- is red color(cname- for different color)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression on test and predicted data')
plt.show()
```

here data points are ploted using actual test dataset and red regression line ploted by using  actual x_test and prdicted y_pred datasets 

by seeing the graph we can say model is fitted littlebit good
